```python
 # using reg we will remove the columns having the (* ,or the ratings in our society columns )
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★','', str(name)).strip()).str.lower()     
```

```python
df['society'] =df['society'].str.replace('nan','Independent')
```

```python
df =df[df['price']!='Price on Request']
```

```python
# writing a function to convert the lakh into crore 
def converting_lakh_into_crore(X):
    if type(X) ==float:
        # if isinastance(X,float)
        return X
    else:
        if X[1] =='Lac':
            return (round(float(X[0])/100,2))
                    
        else:
            return(round(float(X[0]) ,2))
```

```python
df['price'] =df['price'].str.split(' ').apply(converting_lakh_into_crore)
```

```python
df.sample(4)
```

```python
df['price_per_sqft'].value_counts()
```

```python
# Now Dealing the price_per_sqft column in this we will remove the unwanted values 
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹' ,'').str.replace(',','').str.strip().astype('float')
df['price_per_sqft']
```
