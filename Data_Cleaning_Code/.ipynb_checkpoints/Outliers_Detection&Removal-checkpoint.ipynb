- First we will perform the outliers detection on the Numerical Columns 
- Price is our first numerical column
- Steps to perform outliers detection is as folllows
- Check the column ur working with is having normal distribution or not
- If Normal then we need to perform certain steps
- Like use the 



```python
# Plot the displot to check the distribution 
sns.distplot(df['price'])
```

Observations 
- We can see that data points are highly right skewed 
- In this case we need to plot boxplot to check for outliers 

```python
sns.boxplot(data=df ,x='price')
```

Observations |
- With the boxplot it seems that there are a lot outliers
- Now we will try to look into the number of outliers  and then decide are they really outliers or any data error is there 

```python
# Finding the value that are inside IQR and 3 standard deviation 

Q1 =df['price'].quantile(0.25)
Q3 =df['price'].quantile(0.75)
IQR = Q3 -Q1

mean =df['price'].mean()
std  =df['price'].std()

# Finding the Boundary 
upper_limit = (Q3 + 1.5*IQR) 
lower_limit =(Q1 -1.5*IQR)

outliers =df[(df['price']> upper_limit) | (df['price']<lower_limit)]

outliers_value =outliers['price'].describe()
number_of_values =outliers.shape[0]

outliers_value ,number_of_values

```

Observations 
- Here we can see that we have total number of 431 outliers value 
- But before treating them as outliers we should check if these values are actually outliers or there lies any data error under the hood 

```python
outliers[outliers['price']== 31.500000]
```

```python
outliers.head()
```

```python
outliers.to_csv('Outliers.csv')
```

Observation 
- We are having some outliers as well as some data error in our dataset but we r not sure which and how to treat outliers so 
- We will move to next column 

```python
# Price_Per_SQFT
```

```python
sns.displot(df['price_per_sqft'] ,kind ='kde')
# Figure shows that our data points are right skewed not normal so we need use boxplot 
```

```python
sns.boxplot(data=df ,x ='price_per_sqft')
#sns.boxplot(df['price_per_sqft'])
```

Observations 
- We are having  a lot of outliers in this data as we can see
- But always remeber instead of outliers it can also be data error so we must check our dataset also before deciding on outliers
- As we can see that our wihskers are really close to each other and also our graph is not that wide spread

```python
# FINDING THE RANGE 
Q1 = df['price_per_sqft'].quantile(0.25)
Q3 =df['price_per_sqft'].quantile(0.75)
IQR =Q3 -Q1 

# Finding the range as this is not a normally distributed data we will use 
# Formula = Q1 - 1.5*IQR , Q3+1.5*IQR  
upper_limit = Q3 + 1.5*IQR  
lower_limit = Q1 - 1.5*IQR

# Now check for the number of outliers in this column 
outliers = df[(df['price_per_sqft']>upper_limit) | (df['price_per_sqft']<lower_limit)]
number_of_outliers =outliers.shape[0]
outliers_value =outliers['price_per_sqft'].describe()

number_of_outliers ,outliers_value
```

Observations  
- After going through the data we found out that there lies some error in our area column 
- Some values are still in the sq_yard 
- Now we should fix that 

```python
outliers['area'] =outliers['area'].apply(lambda x : x*9 if x<1000 else x )
```

```python
outliers[outliers['area']<1000]
```

```python
outliers['price_per_sqft'] = round(outliers['price']*10000000 /outliers['area'])
```

```python
outliers['price_per_sqft'].describe()
```

```python
df.update(outliers)
```

```python
sns.displot(df['price_per_sqft'])
```

```python
sns.boxplot(df['price_per_sqft'])
```

Observation 
- We can see that our graphs are now improved a lot
